# PIE CHART

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('iris.csv')
data = data.drop(columns =['Id'])
data = data.set_index('Species')
a = list(data.columns)
a = a[0:4]
plt.figure(figsize= (6,8))
plt.pie(data.mean(0),labels =a, radius=1.5,autopct ="%2.1f%%")
plt.legend()
plt.show()
```

```python
a = list(data.columns)
a = a[0:4]
plt.pie(data.mean(0),labels =a, explode=(0.1,0,0.1,0.3), radius=1.5,autopct ="%2.1f%%")
```

```python
a = list(data.columns)
a = a[0:4]
b=data[0:50]
plt.pie(b.mean(0), labels=a, radius=1.5,autopct ="%2.1f%%")
```
